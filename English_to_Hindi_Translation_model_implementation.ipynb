```python
# To get output fast we'll use:
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM
model = AutoModelForSeq2SeqLM.from_pretrained("Helsinki-NLP/opus-mt-en-hi")


```

```python
tokenizer = AutoTokenizer.from_pretrained("Helsinki-NLP/opus-mt-en-hi")
```

```python
model.config.output_attentions
```

```python
print(tokenizer.supported_language_codes)
```

```python
article_en = "We encourages wearing masks and maintain safety"
```

```python
model_inputs = tokenizer.encode(article_en, return_tensors="pt")
```

```python
# translate from English to Hindi
generated_tokens = model.generate(model_inputs)
model_output = tokenizer.decode(generated_tokens[0], skip_special_tokens=True)
print(model_output)




```
